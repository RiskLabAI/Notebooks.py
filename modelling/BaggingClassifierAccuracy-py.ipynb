# Bagging Classifier Accuracy

```python
n = 100
p = np.linspace(0.2, 0.8, n)
N = np.array(range(1, n + 1))

result = np.zeros((n, n))
for (j, p_) in enumerate(p):
    for (i, N_) in enumerate(N):
        result[i, j] = bagging_classifier_accuracy(N_, p_, k=2)
        
```

```python
n = 100
rho = np.linspace(0.0, 1.0, n)
N = np.array(range(5, n + 5))
sigma = 1

def calculate_standard_deviation(n, rho, sigma):
    return sigma**2 * (rho + ((1-rho)/n))

result = np.zeros((n, n))
for (j, rho_) in enumerate(rho):
    for (i, N_) in enumerate(N):
        result[i, j] = calculate_standard_deviation(N_, rho_, sigma=1)
        
import plotly.express as px

fig = px.imshow(result, x=rho, y=N, aspect="auto", width=800, height=800)

fig.update_layout(
    title="Standard Deviation of the bagged prediction (function of ρ and N)",
    xaxis_title="ρ (Average correlation amang forcasts)",
    yaxis_title="N (Number of bagged estimators)",
    template="plotly_dark",
)

fig.write_image("./Figs/Standard Deviation of the bagged prediction.png",format='png',engine='kaleido')
```
