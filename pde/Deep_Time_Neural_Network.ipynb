```python
import RiskLabAI
```

```python
import torch
import numpy as np
from torch import nn, Tensor
config = {
  "dim": 100,
  "total_time": 1,
  "num_time_interval": 50
}
pde = RiskLabAI.pde.PricingDefaultRisk(config)
device = torch.device('cuda')
lossres = []
initres = []


fastsolver = RiskLabAI.pde.FBSDESolver(pde,[150,400,500,400,150],0.001,'Basic')
losses , inits = fastsolver.solve(500,128,(40 + torch.rand(1)*10).to(device),device)
basicloss = losses
```

```python

fastsolver = RiskLabAI.pde.FBSDESolver(pde,[150,400,500,400,150],0.001,'Monte-Carlo')
losses , inits = fastsolver.solve(500,128,(40 + torch.rand(1)*10).to(device),device)
montecarlocloss = losses
```

```python

fastsolver = RiskLabAI.pde.FBSDESolver(pde,[150,400,500,400,150],0.001,'Deep-Time-SetTransformer')
losses , inits = fastsolver.solve(500,128,(40 + torch.rand(1)*10).to(device),device)
settransformerloss = losses
```

```python
torch.tensor(basicloss).cpu().numpy()
```

```python
epochs = list(range(1,501,1))
np_basic_loss = torch.tensor(basicloss).cpu().numpy()
np_montecarlo_loss = torch.tensor(montecarlocloss).cpu().numpy()
np_settransformer_loss = torch.tensor(settransformerloss).cpu().numpy()
```

```python
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import pandas as pd



fig = go.Figure([
    go.Scatter(
        name='Deep Time Monte-Carlo',
        x=epochs[10:],
        y=np_montecarlo_loss[10:],
        mode='lines',
        line=dict(color='green'),
    ),
    go.Scatter(
        name='Deep Time Set Transformer',
        x=epochs[10:],
        y=np_settransformer_loss[10:],
        mode='lines',
        line=dict(color='rgb(31, 119, 180)'),
    ),
    go.Scatter(
        name='Basic',
        x=epochs[10:],
        y=np_basic_loss[10:],
        mode='lines',
        line=dict(color='red'),
    )
])
fig.update_layout(
    yaxis_title='Loss',
    xaxis_title = 'Number of Iteration',
    title='',
    hovermode="x"
)
fig.show()
```
