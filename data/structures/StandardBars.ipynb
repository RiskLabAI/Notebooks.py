# Volume Bars

```python
from controller.data_structure_controller import Controller
import pandas as pd
import numpy as np

controller = Controller()
data = pd.read_csv('../ahrom_edited.csv',)

volume = controller.handle_input_command(
    method_name="volume_standard_bars",
    method_arguments={
        "threshold": 1_000_000,
    },
    input_data=data,
    batch_size=1_000_000
)

# volume = volume.set_index(["Date Time"])
volume
```

```python
volume_imbalance = (volume["Cumulative Buy Volume"] - volume["Cumulative Sell Volume"]).abs()
vpin = volume_imbalance.rolling(window=1).mean() / volume["Cumulative Volume"]
volume["VPIN"] = vpin
volume
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=1)
fig.add_trace(
    go.Scatter(
        x=volume.index,
        y=volume['Close'],
        name="Prices",
    ),
    row=1, col=1,
)
fig.add_trace(
    go.Scatter(
        x=volume.index,
        y=volume['VPIN'],
        name="VPIN",
    ),
    row=1, col=1,
)
fig.update_layout(
    title="Price Prediction",
    xaxis_title="Date",
    hovermode="x unified",
    template="plotly_dark",
)

fig.show()
```

```python
volume.to_csv('../ahrom_vpin.csv')
```
