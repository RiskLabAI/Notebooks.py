## Bubble Types

Speculative bubbles in stock markets are systematic departures from the fundamental price of an asset. *Schatz, M. and D. Sornette (2020). "Inefficient bubbles and efficient drawdowns in financial markets."* introduces a simple framework for bubble modelling based on three bubble-characteristics:

* A pre-drawdown process, $\tilde{S}$, which serves as a model for bubble dynamics and can incor porate superexponential growth and explosive behavior
* The random time of the drawdown,$\tau_J$ being essentially what has become known as the Minsky moment, the time when instabilities built up in the background materialize. To call on Keynes' analogy of musical chairs, it is the time when the music stops.
* The shape of the drawdown, $X$, which serves as a model for a financial crash. The simplest version of a drawdown is an instant drop in asset price values at $\tau_{J}$.

These three bubble characteristics are linked to describe the full stock price $S$ as
$$
\tag{1}
S_{t}=\tilde{S}_{t}\mathbf{1}_{\{t<\tau_{J}\}}+X_{t}\mathbf{1}_{\{\tau_{J}\leqslant t\}},\quad t\in[0,\infty).
$$
The pre-drawdown process $\tilde{S}$ is understood to show exuberant, excessive behavior, thus we require that:

* $\tilde{S}$ violates market efficiency,

It categorizes speculative bubbles in two different types:

### Type-I bubbles
We can model the risk and shape of a drawdown $(\tau_{J}, X)$ being caused by an inefficiency, while extraordinary (pre-drawdown) returns ensure that this risk is “appropriately” priced it in the market: in this case,

* the full price process $S$ is an efficient market,
* while both the pre-drawdown process $\tilde{S}$ and the drawdown process $X$ $-$seen in isolation$-$ are inefficient.

In a type-I bubble, the asset is overvalued for those who mistake $\tilde{S}$ for the true dynamics of the price process. Projecting the pre-drawdown stock dynamics to final payoffs (ignoring the risk of a drawdown), one overvalues the true payoff of the asset. An omniscient observer, assessing the correct underlying $S$, does not see an overvaluation, only the risk of a drawdown compensated by high returns. A rational, omniscient investor stays invested.

In this project, we are going to focus on Type-I bubbles.

### Type-II bubbles
In order to avoid this issue of postulating market efficiency for $S$ to accommodate rational investors, we can either look at markets that are not populated by rational investors with symmetric information or add limits to arbitrage. In the absence of rational investors going against exuberant development, $\tilde{S}$ is driven by some failure of market efficiency. To arrive at a full-fledged bubble model, we assume that the very cause for this inefficiency disappears/gets resolved at $\tau_{J}$ and price levels collapse to “appropriate” levels:

* the drawdown process $X$ is an efficient market,
* while the full process $S$ is inefficient.

In a type-II bubble, the asset is overvalued, even if the true dynamics $S$ are known $- X$ represents the correct valuation of the final payoff. The omniscient observer sees a bubble and, consequently, the risk of a drawdown as possible revaluation. A rational, omniscient investor would exit the market eventually, depending on the likelihood of possible crash times.

## Rational Bubbles

Following *Blanchard, O. J., and M. W. Watson. (1982). “Bubbles, Rational Expectations, and Financial Markets.”* or *Campbell, J. Y., A. W. Lo, and A. C. MacKinlay. (1997). "The Econometrics of Financial
Markets."*, the fundamental price of the asset is derived from the following standard no arbitrage condition:

$$
\tag{2}
P_t=\frac{E_t\left[P_{t+1}+D_{t+1}\right]}{1+R}
$$

where $P_t$ denotes the stock price at period t, $D_{t+1}$ is the dividend for period t, $R$ is the constant risk-free rate, and $E_{t}[.]$ denotes the expectation conditional on the information at time t. Solving Equation (2) by forward iteration yields the fundamental price

$$
\tag{3}
P_{t}^{f}=\sum_{i=1}^{\infty}\frac{1}{(1+R)^{i}}\:E_{t}[D_{t+i}].
$$

Equation (3) states that the fundamental price is equal to the present value of all expected dividend payments. Imposing the transversality condition

$$
\tag{4}
\lim_{k\to\infty}E_t\left[\frac1{(1+R)^k}P_{t+k}\right]=0.
$$

ensures that $P_{t}\:=\:P_{t}^{f}$ is the unique solution of Equation (2) and thereby rules out the existence of a bubble. However, if Equation (4) does not hold, $P_{t}^{f}$is not the only price process that solves Equation (2). Consider a process $\{B_{t}\}_{t=1}^{\infty}$ with the property

$$
\tag{5}
E_{t}\left[B_{t+1}\right]=(1+R)B_{t}.
$$

It can easily be verified that adding $B_{t}$ to $P_{t}^{f}$ will yield another solution to Equation (2). In fact, there are infinitely many solutions. They take the form

$$
\tag{6}
P_{t}=P_{t}^{f}+B_{t},
$$

where $\{B_t\}_{t=1}^{∞}$ is a process that satisfies Equation (5). If a bubble is present in the stock price, Equation (5) requires that any rational investor, who is willing to buy that stock, must expect the bubble to grow at rate $R$. A rational investor is willing to buy an “overpriced” stock since she believes that through price increases, she will be sufficiently compensated for the extra payment $B$.

An obvious problem is that the fundamental component in Equation (6) cannot be directly observed. Therefore, assumptions have to be imposed to characterize the time-series properties of the fundamental price $P_{t}^f$. A convenient and nevertheless empirically plausible assumption is that log dividends follow a random walk with drift

$$
\tag{7}
\log D_t=\mu_D+\log D_{t-1}+q_t\text{,}
$$

where $q_t$ is a white noise process $\mathcal{N}(0, \sigma_D^2)$. Under this assumption, the log fundamental price results as

$$
\tag{8}
P_t^f=\frac{1+G}{R-G}D_t, \\
1+G = e^{\mu_D + \frac{\sigma_D^2}{2}}
$$

(e.g., *Evans, G. W. (1991). "Pitfalls in Testing for Explosive Bubbles in Asset Prices."*). Consequently, if $\log D_{t}$ follows a random walk with drift, so does $\log P_{t}^{f}$ This allows us to distinguish the fundamental price from the bubble process that is characterized by an explosive autoregressive process (see also *Diba, B. T., and H. I. Grossman. (1988). "Explosive Rational Bubbles in Stock Prices?"*).


The crucial condition for rational bubbles is given by Equation (5). The simplest example of a process that satisfies Equation (5) is the deterministic bubble, given by $B_t = (1 + R)^t B_0$, where $B_0$ is an initial value. A somewhat more realistic example, in which the bubble does not necessarily grow forever is taken from *Blanchard, O. J., and M. W. Watson. 1982. “Bubbles, Rational Expectations, and Financial Markets.”*.
$$
\tag{9}
B_{t+1}=\begin{cases}\pi^{-1}(1+R)B_t+\mu_{t+1},& \text{ with probability } \pi \\\\\mu_{t+1}, & \text{ with probability } 1 - \pi\end{cases}
$$

where $\{\mu_t\}_{t=1}^\infty$ of iid random variables with zero mean.

Not every process that satisfies Equation (5) is consistent with rationality For instance, given that stock prices cannot be negative, negative bubbles can be excluded: applying the law of iterated expectations to Equation (5) yields $E_{t}\left[B_{t+\tau}\right]=(1+R)^\tau B_{t}$. If at some time t, $B_t$ was negative, then, as $\tau$ goes to infinity, the expected bubble tends to minus infinity, implying a negative stock price at some future time period. Furthermore, *Diba and Grossman (1988)* argue that a bubble process cannot start from zero. Assume that $B_{t}=0$ at some time $t$ Then, $E_{t}\left[B_{t+1}\right]= 0$, by Equation (5). Assuming nonnegative stock prices, we have $B_{t+1}\geq0$. Together this implies that $B_{t+1}=0$ almost surely.

Although rational bubbles cannot start from zero, they can take a constant
positive value for some time and start to grow exponentially. In his critique of *Diba and Grossman’s (1988)* testing approach, *Evans (1991)* proposed the following model for periodically collapsing bubbles:

$$
\tag{10}
B_{t+1}=\begin{cases}(1+R)B_tu_{t+1} &  \text{ if } B_t \leq \alpha \\\\
\left[\delta+\pi^{-1}(1+R)\theta_{t+1}(B_t-(1+R)^{-1}\delta)\right]u_{t+1}& \text{otherwise} \end{cases}
$$

Here, $\delta$ and $\alpha$ are parameters satisfying $0 < \delta < (1 + R)\alpha$, and $\{u_t\}_{t=1}^{∞}$ is an iid process with $u_t > 0$ and $E_{t}\left[u_{t+1}\right] = 1$, for all $t$. $\{\theta_t\}_{t=1}^{∞}$ is an iid Bernoulli process, where the probability that $\theta_t = 1$ is $\pi$ and the probability that $\theta_t = 0$ is $1 - \pi$,
with $0 < \pi \leq 1$.

## Rational Bubbles on Log Prices

If we assume the risk-free rate is continously compunded and the dividends are paid on log prices, the log fundamental price of the asset is derived from the following standard no arbitrage condition:

$$
\tag{11}
\log P_t=E_t\left[\log P_{t+1}+D_{t+1}-R\right]
$$

where $log P_t$ denotes the log stock price at period $t$, $D_{t+1}$ is the dividend for period t, $R$ is the constant risk-free rate, and $E_{t}[.]$ denotes the expectation conditional on the information at time t. Solving Equation (11) by forward iteration yields the fundamental price

$$
\tag{12}
\log P_{t}^{f}=\sum_{i=1}^{\infty}E_{t}[D_{t+i}].
$$

Equation (12) states that the fundamental price is equal to the present value of all expected dividend payments. Imposing the transversality condition

$$
\tag{13}
\lim_{k\to\infty}E_t\left[\log P_{t+k} - kR\right]=0.
$$

ensures that $P_{t}\:=\:P_{t}^{f}$ is the unique solution of Equation (11) and thereby rules out the existence of a bubble. However, if Equation (13) does not hold, $P_{t}^{f}$is not the only price process that solves Equation (11). Consider a process $\{B_{t}\}_{t=1}^{\infty}$ with the property

$$
\tag{14}
E_{t}\left[B_{t+1}\right]=B_{t}.
$$

It can easily be verified that adding $B_{t}$ to $P_{t}^{f}$ will yield another solution to Equation (2). In fact, there are infinitely many solutions. They take the form

$$
\tag{15}
\log P_{t}= \log P_{t}^{f}+B_{t},
$$

where $\{B_t\}_{t=1}^{∞}$ is a process that satisfies Equation (14).

## Random Walk

To model the $P_{t}^{f}$ random walk process we are going to use the basic Heston model (see *Heston, Steven L. (1993). "A closed-form solution for options with stochastic volatility with applications to bond and currency options"*). The basic Heston model assumes that $S_t$, the price of the asset, is determined by a stochastic process,

$$
\tag{16}
dS_{t}=\mu S_{t}dt+\sqrt{\nu_{t}}S_{t}dW_{t}^{S},
$$

where $\nu_{t}$, the instantaneous variance, is given by a Feller square-root or CIR process,

$$
\tag{17}
d\nu_{t}=\kappa(\theta-\nu_{t})\:dt+\xi\sqrt{\nu_{t}}\:dW_{t}^{\nu}\:,
$$

and $W_{t}^{\mathcal{S}},W_{t}^{\nu}$ are Wiener processes (i.e., continuous random walks) with correlation $\rho$. The model has five parameters:

* $\nu_{0}$, the initial variance.
* $\theta$, the long variance, or long-run average variance of the price; as $t$ tends to infinity, the expected value of $\nu_t$ tends to $\theta$.
* $\rho$, the correlation of the two Wiener processes.
* $\kappa$, the rate at which $\nu_t$ reverts to $\theta$.
* $\xi$, the volatility of the volatility, or 'vol of vol', which determines the variance of $\nu_t$.

lf the parameters obey the $2\kappa\theta>\xi^{2}$ condition (known as the Feller condition) then the process is strictly positive.

# Merton Jump Diffusion Model

The Merton Jump Diffusion model augments the geometric Brownian motion assumption of the Black-Scholes model with an additional jump component. The stock price dynamics is represented as:

$$ dS_t = \mu S_t dt + \sigma S_t dW_t + S_t dJ_t $$

Where:

1. **Drift Term**: $ \mu S_t dt $
   - $ \mu $: Expected return of the stock.
   - Represents the anticipated growth of the stock in the absence of any random shocks or jumps.

2. **Diffusion Term**: $ \sigma S_t dW_t $
   - $ \sigma $: Volatility of the stock.
   - $ dW_t $: Increment of a standard Wiener process (Brownian motion).
   - Captures the continuous random fluctuations of the stock price.

3. **Jump Term**: $ S_t dJ_t $
   - $ dJ_t $: Represents the jump process and captures the abrupt jumps in the stock price.

The jump process $ J_t $ is described as a compound Poisson process:

$$ J_t = \sum_{i=1}^{N(t)} Y_i $$

Where:

- $ N(t) $ is a Poisson process with intensity $ \lambda $. This process dictates the timing of jumps.
- $ Y_i $ are the logarithmic jump sizes, assumed to be i.i.d and follow a normal distribution with mean $ m $ and standard deviation $ s $. This provides the magnitude and direction (upward or downward) of each jump.

Here's a breakdown of the processes:

- **Wiener Process (Brownian motion)**: $ dW_t $
  - Represents the continuous random perturbations in the stock price.
  
- **Poisson Process**: $ N(t) $
  - Governs the timing of jumps. If $ \lambda dt $ is the expected number of jumps in a small interval $ dt $, then the probability of observing one jump in $ dt $ is $ \lambda dt $.

- **Jump Sizes**: $ Y_i $
  - Upon the occurrence of a jump (as per the Poisson process), the magnitude and direction of that jump are determined by a value drawn from the distribution of $ Y_i $, which is normal with mean $ m $ and standard deviation $ s $.

When simulating paths of $ S_t $ using this model:

1. Evolve the stock price considering the drift and diffusion terms.
2. Determine the occurrence of jumps based on the Poisson process.
3. If a jump occurs, adjust the stock price by the jump magnitude from the distribution of $ Y_i $.

This blend of continuous price evolution with sporadic jumps offers a more nuanced framework that might capture stock price behaviors in the real world more effectively than the standard geometric Brownian motion.


\subsection*{The Drift Burst Hypothesis}
In an article, \textit{K. Christensen, R. Oomen, and R. Reno. "The drift burst hypothesis."} postulate the so-called drift burst hypothesis to model short-lived flash crashes in high-frequency tick data.

They delve into the intricacies of financial market processes, emphasizing the critical interplay between drift and volatility. They postulate that explosive drift can only be sustainable when accompanied by similar behavior in volatility. Central to their exploration is the concept of a "volatility burst", characterized by a sudden spike in market volatility. The authors provide a mathematical formulation to capture this phenomenon, linking the magnitude of the volatility burst to specific market conditions. Their findings suggest that traditional notions of market behavior, especially concerning arbitrage opportunities, may need revisiting, given the complex dynamics they outline.

Central to their discourse are several foundational assumptions about the market process:

\begin{itemize}
    \item \textbf{Stochastic Differential Equations (SDEs):} The authors model the financial market processes using SDEs. These equations play a crucial role in quantifying the market dynamics, allowing them to describe complex behaviors in the market, especially in the presence of random shocks.
    \begin{equation}
        \mathrm{d}X_t = \mu_t \mathrm{d}t + \sigma_t \mathrm{d}W_t
    \end{equation}
    
    \item \textbf{Drift and Volatility Dynamics:} The researchers emphasize the relationship between drift (the expected change in the asset price) and volatility (the unpredictability of the price change). They propose that while the drift can show explosive tendencies, such a scenario is only tenable if it's balanced by a corresponding explosion in volatility.
    
    \item \textbf{Semimartingale Property:} They underline that despite having an explosive drift as defined in their formulations, the process $X$ they refer to remains a semimartingale. This property ensures that the process can still be integrated and remains within the scope of conventional stochastic calculus.
    
    \item \textbf{Volatility Burst:} Another innovative concept introduced is that of a "volatility burst". They model it using a mathematical formulation where the volatility of an asset shows a sudden and extreme increase. The magnitude and behavior of this burst are related to specific market conditions, and it serves as an essential aspect of their market dynamics model.
\end{itemize}

The no-arbitrage condition is pivotal in financial mathematics, ensuring that there's no possibility of making a riskless profit. Christensen, Oomen, and Reno delve deep into this topic to argue how their model preserves this condition:

\begin{itemize}
    \item \textbf{Semimartingale \& Arbitrage:} They state that the process $X$, even with a drift as outlined in their assumptions, remains a semimartingale. This property is essential, but not sufficient on its own, to exclude arbitrage from the model. The semimartingale condition is a necessary criterion for ensuring that the process doesn't provide trivial arbitrage opportunities. The process $X$ is a semimartingale, which implies:
    \begin{equation}
        X_t = X_0 + M_t + A_t
    \end{equation}
    Where $M_t$ is a martingale and $A_t$ is an adapted and predictable process.
    
    \item \textbf{Girsanov's Theorem \& Equivalent Martingale Measure:} The authors allude to Girsanov's Theorem, a fundamental result in the realm of financial mathematics. It sets a necessary and sufficient condition for the existence of an equivalent martingale measure, a critical concept ensuring the no-arbitrage property of a financial market.
    
    \item \textbf{Structural Condition:} The authors introduce a "structural condition", mathematically represented as an integral containing a ratio of drift to volatility. This condition essentially checks for the potential of "free lunch with vanishing risk". If the drift explodes in a specific way without a corresponding increase in volatility, then the model could provide an opportunity for arbitrage. Thus, they contend that the explosive nature of volatility is a required condition to ensure that bursts in drift do not introduce arbitrage opportunities. The no-arbitrage condition is ensured by:
    \begin{equation}
        \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}\left(\frac{\mu_s}{\sigma_s}\right)^2 \mathrm{~d} s < \infty
    \end{equation}
    This condition avoids a "free lunch with vanishing risk."
\end{itemize}

In essence, through intricate mathematical reasoning and market dynamics modeling, Christensen, Oomen, and Reno establish that their presented model, even with its innovative assumptions and behaviors, still adheres to the foundational principle of no-arbitrage in financial markets.

\section*{Fixed Point $ \tau_{\mathrm{db}} $ and the Behavior of $ \mu $ and $ \sigma $}

The authors pinpoint a specific moment in time, $ \tau_{\mathrm{db}} $, which is of interest when studying the drift $ \mu $ and volatility $ \sigma $ behaviors. Under above assumptions, both $ \mu $ and $ \sigma $ are described as locally bounded. This boundedness has the following implications as $ \bar{\Delta} $ approaches 0:

\begin{itemize}
    \item The integrated absolute drift over a time interval $\bar{\Delta}$ centered around $ \tau_{\mathrm{db}} $ grows linearly with $ \bar{\Delta} $. Specifically,
    \begin{equation}
    \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}|\mu_s| \mathrm{d} s = O_p(\bar{\Delta})
    \end{equation}

    \item The integrated volatility over the same interval grows with the square root of $ \bar{\Delta} $. Specifically,
    \begin{equation}
    \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}} \sigma_s \mathrm{d} W_s = O_p(\sqrt{\bar{\Delta}})
    \end{equation}
\end{itemize}

From these observations, one can deduce that as $ \bar{\Delta} $ shrinks, the drift's impact is overshadowed by the volatility. In essence, the drift becomes negligible compared to the volatility in very short time intervals. This is because $ \bar{\Delta} $ grows at a much slower rate than $ \sqrt{\bar{\Delta}} $. This understanding forms the foundation for econometricians' general disregard for the drift when analyzing market data, as the volatility seems to play a far more dominating role over short periods.

\section*{Drift Burst and Its Significance}

However, the paper proposes an alternative market model where the drift is allowed to explode in the vicinity of $ \tau_{\mathrm{db}} $. In this alternative scenario:
\begin{equation}
\int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}|\mu_s| \mathrm{d} s = O_p(\bar{\Delta}^{\gamma_\mu})
\end{equation}
with a condition $ 0 < \gamma_\mu < 1/2 $.

This exploding drift is termed as a 'drift burst', and the point $ \tau_{\mathrm{db}} $ is referred to as the 'drift burst time'. The $ \gamma_\mu > 0 $ condition ensures the continuity of $ X $, the underlying stochastic process.

The concept of a "drift burst" is introduced to capture extreme and sudden movements in the drift of a financial asset. This abrupt change can lead to significant price fluctuations in a very short time span.

\begin{itemize}
    \item \textbf{Nature of Drift Burst}:
    \begin{itemize}
        \item Drift bursts occur when the drift term in a stochastic differential equation (SDE) displays an explosive behavior, especially around a certain point in time, denoted as $ \tau_{\mathrm{db}} $.
        \item This sudden change is modeled in a way that as the time $ t $ approaches $ \tau_{\mathrm{db}} $, the drift explodes, thereby causing erratic price movements.
    \end{itemize}
    
    \item \textbf{Characteristics}:
    \begin{itemize}
        \item The drift burst behavior is characterized by a particular parameter $ \alpha $ which determines the intensity of the burst. The model restricts the value of $ \alpha $ to ensure mathematical feasibility and relevance.
        \item The amplitude or the magnitude of the drift burst is controlled by another parameter, $ a $, which is positive and signifies the severity of the drift movement.
    \end{itemize}
    
    \item \textbf{Economic Interpretation}:
    \begin{itemize}
        \item A drift burst can be thought of as a sudden news event or a market shock that leads to drastic price changes. These events might be sudden regulatory changes, geopolitical incidents, or any other unforeseen major market-moving news.
    \end{itemize}
    
    \item \textbf{Expression for Drift Burst}:
    \begin{equation}
    \mu_t^{\mathrm{db}} = a |\tau_{\mathrm{db}} - t|^{-\alpha}
    \end{equation}
    where:
    \begin{itemize}
        \item $ \mu_t^{\mathrm{db}} $ represents the drift at time $ t $ with the drift burst behavior.
        \item $ a $ is a positive parameter representing the amplitude of the drift burst.
        \item $ \alpha $ is the intensity of the burst, with the condition $ 1/2 < \alpha < 1 $ ensuring feasibility.
    \end{itemize}
\end{itemize}

\section*{Volatility Burst and Its Relevance}

Similarly, the concept of a 'volatility burst' is introduced. A volatility burst happens if:
\begin{equation}
\int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}} \sigma_s \mathrm{d} W_s = O_p(\bar{\Delta}^{\gamma_\sigma})
\end{equation}
where $ 0 < \gamma_\sigma < 1/2 $.

Volatility bursts capture the extreme and abrupt changes in the volatility or the variability of an asset price. A sudden spike in volatility can increase the risk and uncertainty in the market.

\begin{itemize}
    \item \textbf{Nature of Volatility Burst}:
    \begin{itemize}
        \item A volatility burst happens when the volatility term in the SDE showcases a dramatic increase, especially around a specific time, denoted as $ \tau_{\mathrm{db}} $, which is the same as the time of the drift burst.
        \item The reason for tying both drift and volatility bursts to the same time point is to capture the idea that significant market events usually lead to both price and volatility spikes.
    \end{itemize}
    
    \item \textbf{Characteristics}:
    \begin{itemize}
        \item The behavior of the volatility burst is governed by the parameter $\beta $, which indicates the intensity of the volatility movement. For the model to make economic and mathematical sense, $ \beta $ is restricted to a certain range.
        \item The magnitude of the volatility burst is dictated by the positive parameter $ b $, determining how drastic the change in volatility is.
    \end{itemize}
    
    \item \textbf{Economic Interpretation}:
    \begin{itemize}
        \item Volatility bursts can be thought of as the market's reaction to sudden and unexpected news. As the market tries to assimilate this new information, uncertainty (and hence volatility) rises sharply. This might be due to varying interpretations of the news, liquidity constraints, or simply panic selling/buying.
    \end{itemize}
    
    \item \textbf{Expression for Volatility Burst}:
    \begin{equation}
    \sigma_t^{\mathrm{vb}} = b |\tau_{\mathrm{db}} - t|^{-\beta}
    \end{equation}
    where:
    \begin{itemize}
        \item $ \sigma_t^{\mathrm{vb}} $ represents the volatility at time $ t $ with the volatility burst behavior.
        \item $ b $ is a positive parameter indicating the severity of the volatility movement.
        \item $ \beta $ is the intensity of the burst, with the condition $ 0 < \beta < 1/2 $.
    \end{itemize}
\end{itemize}

The constraint on $ \beta $ ensures the feasibility of the model by making certain the integrability of $ \sigma^{\mathrm{vb}} $. Specifically, the following must hold:
\begin{equation}
\int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}(\sigma_s^{\mathrm{vb}})^2 \mathrm{d} s < \infty
\end{equation}

The relationship between $ \gamma_\sigma $ and $ \beta $ is further clarified:
\begin{equation}
\gamma_\sigma = 1/2 - \beta
\end{equation}

This in-depth look into both the drift and volatility bursts offers insights into scenarios where the traditional assumptions (of drift being negligible over short periods) may not hold.

Finally, for equation (22) to hold, we should have:
\begin{equation}
\alpha-\beta < 1/2
\end{equation}

Thus, they contend that the explosive nature of volatility is a required condition to ensure that bursts in drift do not introduce arbitrage opportunities. The no-arbitrage condition is ensured by:
  $$
  \tag{22}
  \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}\left(\frac{\mu_s}{\sigma_s}\right)^2 \mathrm{~d} s < \infty
  $$
  This condition avoids a "free lunch with vanishing risk."


In essence, through intricate mathematical reasoning and market dynamics modeling, Christensen, Oomen, and Reno establish that their presented model, even with its innovative assumptions and behaviors, still adheres to the foundational principle of no-arbitrage in financial markets.

### Fixed Point $ \tau_{\mathrm{db}} $ and the Behavior of $ \mu $ and $ \sigma $

The authors pinpoint a specific moment in time, $ \tau_{\mathrm{db}} $, which is of interest when studying the drift $ \mu $ and volatility $ \sigma $ behaviors. Under above assumptions, both $ \mu $ and $ \sigma $ are described as locally bounded. This boundedness has the following implications as $ \bar{\Delta} $ approaches 0:
- The integrated absolute drift over a time interval $\bar{\Delta}$ centered around $ \tau_{\mathrm{db}} $ grows linearly with $ \bar{\Delta} $. Specifically,
$$
\tag{25}
\int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}|\mu_s| \mathrm{d} s = O_p(\bar{\Delta})
$$

- The integrated volatility over the same interval grows with the square root of $ \bar{\Delta} $. Specifically,
$$\tag{26} \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}} \sigma_s \mathrm{d} W_s = O_p(\sqrt{\bar{\Delta}}) $$

From these observations, one can deduce that as $ \bar{\Delta} $ shrinks, the drift's impact is overshadowed by the volatility. In essence, the drift becomes negligible compared to the volatility in very short time intervals. This is because $ \bar{\Delta} $ grows at a much slower rate than $ \sqrt{\bar{\Delta}} $. This understanding forms the foundation for econometricians' general disregard for the drift when analyzing market data, as the volatility seems to play a far more dominating role over short periods.

### Drift Burst and Its Significance

However, the paper proposes an alternative market model where the drift is allowed to explode in the vicinity of $ \tau_{\mathrm{db}} $. In this alternative scenario:
$$\tag{27} \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}|\mu_s| \mathrm{d} s = O_p(\bar{\Delta}^{\gamma_\mu}) $$
with a condition $ 0 < \gamma_\mu < 1/2 $.

This exploding drift is termed as a 'drift burst', and the point $ \tau_{\mathrm{db}} $ is referred to as the 'drift burst time'. The $ \gamma_\mu > 0 $ condition ensures the continuity of $ X $, the underlying stochastic process.

The concept of a "drift burst" is introduced to capture extreme and sudden movements in the drift of a financial asset. This abrupt change can lead to significant price fluctuations in a very short time span.

- **Nature of Drift Burst**:
    - Drift bursts occur when the drift term in a stochastic differential equation (SDE) displays an explosive behavior, especially around a certain point in time, denoted as $ \tau_{\mathrm{db}} $.
    - This sudden change is modeled in a way that as the time $ t $ approaches $ \tau_{\mathrm{db}} $, the drift explodes, thereby causing erratic price movements.
    
- **Characteristics**:
    - The drift burst behavior is characterized by a particular parameter $ \alpha $ which determines the intensity of the burst. The model restricts the value of $ \alpha $ to ensure mathematical feasibility and relevance.
    - The amplitude or the magnitude of the drift burst is controlled by another parameter, $ a $, which is positive and signifies the severity of the drift movement.
    
- **Economic Interpretation**:
    - A drift burst can be thought of as a sudden news event or a market shock that leads to drastic price changes. These events might be sudden regulatory changes, geopolitical incidents, or any other unforeseen major market-moving news.
    
- **Expression for Drift Burst**:
    $$ \tag{23} \mu_t^{\mathrm{db}} = a |\tau_{\mathrm{db}} - t|^{-\alpha} $$
    where:
    - $ \mu_t^{\mathrm{db}} $ represents the drift at time $ t $ with the drift burst behavior.
    - $ a $ is a positive parameter representing the amplitude of the drift burst.
    - $ \alpha $ is the intensity of the burst, with the condition $ 1/2 < \alpha < 1 $ ensuring feasibility.

### Volatility Burst and Its Relevance

Similarly, the concept of a 'volatility burst' is introduced. A volatility burst happens if:
$$\tag{28} \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}} \sigma_s \mathrm{d} W_s = O_p(\bar{\Delta}^{\gamma_\sigma}) $$
where $ 0 < \gamma_\sigma < 1/2 $.

Volatility bursts capture the extreme and abrupt changes in the volatility or the variability of an asset price. A sudden spike in volatility can increase the risk and uncertainty in the market.

- **Nature of Volatility Burst**:
    - A volatility burst happens when the volatility term in the SDE showcases a dramatic increase, especially around a specific time, denoted as $ \tau_{\mathrm{db}} $, which is the same as the time of the drift burst.
    - The reason for tying both drift and volatility bursts to the same time point is to capture the idea that significant market events usually lead to both price and volatility spikes.
    
- **Characteristics**:
    - The behavior of the volatility burst is governed by the parameter $\beta $, which indicates the intensity of the volatility movement. For the model to make economic and mathematical sense, $ \beta $ is restricted to a certain range.
    - The magnitude of the volatility burst is dictated by the positive parameter $ b $, determining how drastic the change in volatility is.
    
- **Economic Interpretation**:
    - Volatility bursts can be thought of as the market's reaction to sudden and unexpected news. As the market tries to assimilate this new information, uncertainty (and hence volatility) rises sharply. This might be due to varying interpretations of the news, liquidity constraints, or simply panic selling/buying.

- **Expression for Volatility Burst**:
    $$ \tag{24} \sigma_t^{\mathrm{vb}} = b |\tau_{\mathrm{db}} - t|^{-\beta} $$
    where:
    - $ \sigma_t^{\mathrm{vb}} $ represents the volatility at time $ t $ with the volatility burst behavior.
    - $ b $ is a positive parameter indicating the severity of the volatility movement.
    - $ \beta $ is the intensity of the burst, with the condition $ 0 < \beta < 1/2 $.

The constraint on $ \beta $ ensures the feasibility of the model by making certain the integrability of $ \sigma^{\mathrm{vb}} $. Specifically, the following must hold:
$$\tag{29} \int_{\tau_{\mathrm{db}}-\bar{\Delta}}^{\tau_{\mathrm{db}}+\bar{\Delta}}(\sigma_s^{\mathrm{vb}})^2 \mathrm{d} s < \infty $$

The relationship between $ \gamma_\sigma $ and $ \beta $ is further clarified:
$$ \tag{30} \gamma_\sigma = 1/2 - \beta $$

This in-depth look into both the drift and volatility bursts offers insights into scenarios where the traditional assumptions (of drift being negligible over short periods) may not hold.

Finally, for equation (22) to hold, we should have:
$$
\tag{31}
\alpha-\beta < 1/2
$$






# Markov Chains

Markov chains are defined primarily by their transition probabilities. These probabilities can be concisely represented using a few key equations.

## Transition Probability:

Given a finite number of states $S = \{s_1, s_2, \ldots, s_n\}$, the probability of transitioning from state $s_i$ to state $s_j$ in one step is denoted by $P_{ij}$.

$$
\tag{32}
P_{ij} = P(X_{n+1} = s_j | X_n = s_i)
$$

Where:
- $X_n$ is the state at time $n$.
- $P_{ij}$ is the entry in the $i^{th}$ row and $j^{th}$ column of the transition matrix $P$.

## Transition Matrix:

The matrix $P = [P_{ij}]$ is called the transition matrix of the Markov chain. Each entry $P_{ij}$ represents the one-step transition probability from state $s_i$ to state $s_j$.

$$
P = \begin{bmatrix}
P_{11} & P_{12} & \cdots & P_{1n} \\
P_{21} & P_{22} & \cdots & P_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
P_{n1} & P_{n2} & \cdots & P_{nn}
\end{bmatrix}
$$

Properties:
1. $0 \leq P_{ij} \leq 1$: Since $P_{ij}$ is a probability.
2. $\sum_{j=1}^n P_{ij} = 1$: The sum of probabilities from a state $s_i$ to all other states (including itself) is 1.

## n-Step Transition Probability:

The probability of transitioning from state $s_i$ to state $s_j$ in $n$ steps is denoted by $P_{ij}^{(n)}$, which can be obtained by raising the transition matrix $P$ to the power $n$.

$$
\tag{33}
P_{ij}^{(n)} = \left( P^n \right)_{ij}
$$

Where:
- $(P^n)_{ij}$ is the entry in the $i^{th}$ row and $j^{th}$ column of the matrix $P^n$.





# Discrete Simulation
In the discrete setting (like when using the Euler-Maruyama method for approximation), this is translated to:

$$ \Delta S_t \approx \mu S_t \Delta t + \sqrt{\nu_t} S_t Z \sqrt{\Delta t} + Y \Delta N(t)$$

where $ Z $ is a standard normal random variable, $Y$ a normal random variable with mean $m$ and variance $v^2$ and $\Delta N(t)$ is the increment of the Poisson process which has an expected number of jumps $\lambda \Delta t$.

$\nu_{t}$, the instantaneous variance, is given by
$$
\tag{19}
\Delta \nu_t =  \kappa(\nu_{t} - \theta) \Delta t + \xi\sqrt{\nu_{t}}\:(\rho_\epsilon\epsilon_t^P + \epsilon^{\nu}_t \sqrt{1 - \rho_\epsilon^2})\: \sqrt{\Delta t}
$$
where $\epsilon^{\nu}_t$ and $\epsilon_t^P$ are standard white noises $\mathcal{N}(0, 1)$. The noise of two processes exhibit a correlation of $\rho_\epsilon$.
In each simulation, $\nu_t$ is initiated at random from its stationary law $v_t \sim \operatorname{Gamma}\left(2 \kappa \theta \xi^{-2}, 2 \kappa \xi^{-2}\right)$.

So, $ \sqrt{dt} $ is a scaling factor that comes from the properties of the Brownian motion increment when approximating the continuous stochastic process in a discrete setting.

To clarify, the log return over a small interval $\Delta t$ is approximately:
$$
r_{t+\Delta t}=\ln \left(\frac{S_{t+\Delta t}}{S_t}\right)
$$
Using the Ito's Lemma, we can derive the following for log returns:
$$
r_{t+dt}=\left(\mu-\frac{1}{2} \nu_t -\lambda\left(m+\frac{v^2}{2}\right) \right) d t+\sqrt{\nu_t} d W_t^S + Y_t dN(t)
$$


```python
import numpy as np
from numpy.random import standard_normal, gamma
from numba import jit, prange

@jit(nopython=True)
def initialize_volatility(kappa, theta, xi, n_simulations):
    """
    Initialize the variance volatility using its stationary distribution.
    """
    shape = 2 * kappa * theta / xi**2
    scale = xi**2 / (2 * kappa)
    return gamma(shape, scale, n_simulations)

@jit(nopython=True, parallel=True)
def simulate_heston(
    num_steps: int,
    mu: float,
    kappa: float,
    theta: float,
    xi: float,
    rho_epsilon: float,
    n_simulations: int
) -> np.ndarray:
    r"""
    Simulate the Heston model for a given set of parameters.

    :param num_steps: Number of steps in the time horizon.
    :param mu: Expected return.
    :param kappa: Rate of reversion.
    :param theta: Long-term mean of the variance.
    :param xi: Volatility of the instantaneous variance.
    :param rho_epsilon: Correlation between the two Wiener processes.
    :param n_simulations: Number of paths to simulate.

    :return: np.ndarray:
             - P: Simulated paths of the asset price.

    Mathematical details:

    .. math::
        dS_{t}=\mu S_{t}dt+\sqrt{\nu_{t}}S_{t}dW_{t}^{S}

    .. math::
        d\nu_{t}=\kappa(\theta-\nu_{t})\:dt+\xi\sqrt{\nu_{t}}\:dW_{t}^{\nu}
    """
    # Preallocate arrays for asset and variance paths
    log_price = np.zeros((n_simulations, num_steps+1))
    volatility = np.zeros((n_simulations, num_steps+1))

    # Set initial values
    log_price[:, 0] = 1  # Initial asset log price
    volatility[:, 0] = initialize_volatility(kappa, theta, xi, n_simulations)

    for i in prange(n_simulations):
        for t in range(1, num_steps + 1):
            # Sample random numbers
            eps_p = standard_normal()
            eps_nu = rho_epsilon * eps_p + np.sqrt(1 - rho_epsilon**2) * standard_normal()

            # Update volatility (variance) process
            volatility[i, t] = (1 - kappa) * volatility[i, t - 1] + kappa * theta + xi * np.sqrt(volatility[i, t - 1]) * eps_nu

            # Ensure non-negativity of volatility
            volatility[i, t] = max(volatility[i, t], 0)

            # Update asset price process
            log_return = mu + np.sqrt(volatility[i, t - 1]) * eps_p
            log_price[i, t] = log_price[i, t - 1] * np.exp(log_return)

    return log_price, volatility

import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_steps = 1000
mu = 0.0004
kappa = 5
theta = 0.0001
xi = 0.04
rho_epsilon = -0.5
num_simulations = 10

# Simulate using the Heston model
asset_paths, variances = simulate_heston(num_steps, mu, kappa, theta, xi, rho_epsilon, num_simulations)

# Plot the first 5 asset paths
plt.figure(figsize=(12,6))
for i in range(num_simulations):
    plt.plot(asset_paths[i], label=f'Simulation {i+1}')
plt.title('Asset Paths Simulated Using the Heston Model')
plt.xlabel('Time Step')
plt.ylabel('Asset Price')
plt.legend()
plt.show()

# Plot the first 5 variance paths
plt.figure(figsize=(12,6))
for i in range(num_simulations):
    plt.plot(variances[i], label=f'Simulation {i+1}')
plt.title('Variance Paths Simulated Using the Heston Model')
plt.xlabel('Time Step')
plt.ylabel('Variance')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from numba import jit
import quantecon as qe
from joblib import Parallel, delayed

@jit(nopython=True)
def compute_log_returns(N, mu_vec, kappa_vec, theta_vec, xi_vec, dwS, dwV, Y, n, dt, sqrtdt, lam_vec, m_vec, v_vec):
    v = np.empty(N+1)
    dJ = np.multiply(n, Y)
    v[0] = theta_vec[0]
    log_returns = np.empty(N)
    for i in range(N):
        v[i+1] = v[i] + kappa_vec[i] * (theta_vec[i] - max(v[i], 0)) * dt + xi_vec[i] * np.sqrt(max(v[i], 0)) * dwV[i] * sqrtdt
        log_returns[i] = (mu_vec[i] - 0.5 * v[i] - lam_vec[i] * (m_vec[i] + (v_vec[i] ** 2) / 2)) * dt + np.sqrt(v[i]) * dwS[i] * sqrtdt + dJ[i]

    return log_returns

def heston_merton_log_returns(T, N, mu_vec, kappa_vec, theta_vec, xi_vec, rho_vec, lam_vec, m_vec, v_vec, random_state=None):
    # Ensure the length of parameter vectors is N
    assert len(mu_vec) == len(kappa_vec) == len(theta_vec) == len(xi_vec) == len(rho_vec) == N

    rng = np.random.default_rng(random_state)
    dt = T/N
    sqrtdt = np.sqrt(dt)

    # Generate Gaussian random numbers
    z = np.zeros((N, 3))
    n = np.zeros(N)
    for i in range(N):
        z[i] = rng.multivariate_normal(
            [0, 0, m_vec[i]],
            [[1.0,       rho_vec[i],     0.0],
             [rho_vec[i], 1.0,           0.0],
             [0.0,        0.0, v_vec[i] ** 2]]
        )
        n[i] = rng.poisson(lam_vec[i] * dt)

    dwS = z[:, 0]
    dwV = z[:, 1]
    Y = z[:, 2]

    return compute_log_returns(N, mu_vec, kappa_vec, theta_vec, xi_vec, dwS, dwV, Y, n, dt, sqrtdt, lam_vec, m_vec, v_vec)

def align_params_length(regime_params):
    max_len = max([len(value) if isinstance(value, list) else 1 for value in regime_params.values()])

    for key, value in regime_params.items():
        if isinstance(value, list):
            if len(value) < max_len:
                regime_params[key].extend([value[-1]] * (max_len - len(value)))
        else:
            regime_params[key] = [value] * max_len

    return regime_params, max_len

def generate_prices_from_regimes(regimes, P, T, N, random_state):
    mc = qe.MarkovChain(P, state_values=list(regimes.keys()))
    simulated_regimes = mc.simulate(ts_length=N, random_state=random_state)

    mu_vec, kappa_vec, theta_vec, xi_vec, rho_vec, lam_vec, m_vec, v_vec = [], [], [], [], [], [], [], []

    extension = 0
    for i, regime in enumerate(np.copy(simulated_regimes)):
        params, max_len = align_params_length(regimes[regime].copy())
        mu_vec.extend(params['mu'])
        kappa_vec.extend(params['kappa'])
        theta_vec.extend(params['theta'])
        xi_vec.extend(params['xi'])
        rho_vec.extend(params['rho'])
        lam_vec.extend(params['lam'])
        m_vec.extend(params['m'])
        v_vec.extend(params['v'])

        if max_len > 1:
            simulated_regimes = np.insert(simulated_regimes, i + extension, [regime] * (max_len-1))
            extension += max_len - 1

    # Truncate parameter vectors if they exceed N
    mu_vec = np.array(mu_vec[:N])
    kappa_vec = np.array(kappa_vec[:N])
    theta_vec = np.array(theta_vec[:N])
    xi_vec = np.array(xi_vec[:N])
    rho_vec = np.array(rho_vec[:N])
    lam_vec = np.array(lam_vec[:N])
    m_vec = np.array(m_vec[:N])
    v_vec = np.array(v_vec[:N])
    simulated_regimes = simulated_regimes[:N]

    log_returns = heston_merton_log_returns(T, N, mu_vec, kappa_vec, theta_vec, xi_vec, rho_vec, lam_vec, m_vec, v_vec, random_state=random_state)
    prices = np.exp(np.cumsum(log_returns))

    return prices, simulated_regimes

def parallel_generate_prices(n_paths, regimes, P, T, N, random_state=None, n_jobs=1):
    rng = np.random.default_rng(random_state)
    random_states = rng.choice(10*n_paths, size=n_paths, replace=False)
    results = Parallel(n_jobs=n_jobs)(delayed(generate_prices_from_regimes)(regimes, P, T, N, rs) for rs in random_states)

    prices, simulated_regimes = zip(*results)

    # Convert tuples directly to DataFrames
    prices_df = pd.DataFrame(prices)
    simulated_regimes_df = pd.DataFrame(simulated_regimes)

    return prices_df.T, simulated_regimes_df.T
```

```python
@jit(nopython=True)
def drift_volatility_burst(bubble_length, a_before, a_after, b_before, b_after, alpha, beta):
    steps = np.linspace(0, 1, bubble_length)

    # Handle a values based on step
    a_values = np.where(steps <= 0.5, a_before, a_after)
    b_values = np.where(steps <= 0.5, b_before, b_after)

    # Calculate drifts and volatilities
    denominators = np.abs(steps - 0.5)
    denominators[steps == 0.5] = np.nan  # Set the denominator to nan for step == 0.5

    drifts = a_values / denominators ** alpha
    volatilities = b_values / denominators ** beta

    # Fill nan values with preceding values
    mask = np.isnan(denominators)
    if np.sum(mask) > 0:
        drifts[mask] = 0
        volatilities[mask] = volatilities[np.where(mask)[0][0] - 1]

    return drifts, volatilities

bubble_drifts, bubble_volatilities = drift_volatility_burst(60, 1.5, -1.5, 0.15, 0.15, 0.65, 0.2)
flash_drifts, flash_volatilities = drift_volatility_burst(10, -20, 20, 0.2, 0.2, 0.65, 0.2)
# Dictionary of Heston parameters for different market regimes
regimes = {
    'calm': {
        'mu': 0.2,
        'kappa': 1.5,
        'theta': 0.04,
        'xi': 0.3,
        'rho': -0.3,
        'lam': 2,
        'm': 0,
        'v': 0.08
    },
    'volatile': {
        'mu': 0.2,
        'kappa': 1.0,
        'theta': 0.5,
        'xi': 0.5,
        'rho': -0.7,
        'lam': 3,
        'm': 0,
        'v': 0.15
    },
    'speculative_bubble': {
        'mu': list(bubble_drifts),
        'kappa': 1,
        'theta': list(bubble_volatilities),
        'xi': 0,
        'rho': 0,
        'lam': 0,
        'm': 0,
        'v': 0.00000001
    },
    'flash_crash': {
        'mu': list(flash_drifts),
        'kappa': 1,
        'theta': list(flash_volatilities),
        'xi': 0,
        'rho': 0,
        'lam': 0,
        'm': 0,
        'v': 0.00000001
    },
}

# Example transition matrix for two regimes (calm and volatile)
# Probabilities will depend on your expectations
# P = [[0.95, 0.05],  # transition probabilities from calm
#      [0.1, 0.9]]  # transition probabilities from volatile

P = np.array([
    [0.7,   0.298, 0.001, 0.001],  # State 0 transitions
    [0.298, 0.7,   0.001, 0.001],  # State 1 transitions
    [0.7,   0.3,   0.0,   0.0  ],      # State 2 transitions
    [0.7,   0.3,   0.0,   0.0  ]       # State 3 transitions
])

T = 4
N = T * 252

prices, simulated_regimes = generate_prices_from_regimes(regimes, P, T, N, 10)
```

```python
import numpy as np
from itertools import combinations
from typing import List, Tuple, Callable
from numba import jit
from joblib import Parallel, delayed

@jit(nopython=True)
def performance_evaluation(
    train_partition: np.ndarray,
    test_partition: np.ndarray,
    n_strategies: int,
    metric: Callable,
    risk_free_return: float
) -> Tuple[bool, float]:
    """
    Evaluate the performance of various strategies on given train and test partitions and
    compute the logit value to determine if the best in-sample strategy is overfitting.

    :param train_partition: Training data partition used for evaluating in-sample performance.
    :type train_partition: np.ndarray
    :param test_partition: Testing data partition used for evaluating out-of-sample performance.
    :type test_partition: np.ndarray
    :param n_strategies: Number of strategies to evaluate.
    :type n_strategies: int
    :param metric: Metric function for evaluating strategy performance.
                   The function should accept a data array and risk_free_return as arguments.
    :type metric: Callable
    :param risk_free_return: Risk-free return used in the metric function, often used for Sharpe ratio.
    :type risk_free_return: float

    :return: Tuple where the first value indicates if the best in-sample strategy is overfitting
             (True if overfitting, False otherwise) and the second value is the logit value computed.
    :rtype: Tuple[bool, float]
    """

    evaluate_train = list(map(lambda i: metric(train_partition[:, i], risk_free_return), range(n_strategies)))
    best_strategy = np.argmax(evaluate_train)

    evaluate_test = list(map(lambda i: metric(test_partition[:, i], risk_free_return), range(n_strategies)))

    rank_of_best_is_strategy = np.argsort(evaluate_test).tolist().index(best_strategy) + 1

    w_bar = rank_of_best_is_strategy / (n_strategies + 1)
    logit_value = np.log(w_bar / (1 - w_bar))

    return logit_value <= 0.0, logit_value

def probability_of_backtest_overfitting(
    matrix_data: np.ndarray,
    n_partitions: int,
    metric: Callable,
    risk_free_return: float = 0.0,
    n_jobs: int = 1
) -> Tuple[float, np.ndarray]:
    """
    Computes the Probability Of Backtest Overfitting.

    For instance, if \(S=16\), we will form 12,780 combinations.

    .. math::
        \left(\begin{array}{c}
        S \\
        S / 2
        \end{array}\right) = \prod_{i=0}^{S / 2^{-1}} \frac{S-i}{S / 2-i}

    :param matrix_data: Matrix of T×N for T observations on N strategies.
    :type matrix_data: np.ndarray
    :param n_partitions: Number of partitions (must be even).
    :type n_partitions: int
    :param metric: Metric function for evaluating strategy.
    :type metric: Callable
    :param risk_free_return: Risk-free return for calculating Sharpe ratio.
    :type risk_free_return: float
    :param n_jobs: Number of parallel jobs.
    :type n_jobs: int

    :return: Tuple containing Probability Of Backtest Overfitting and an array of logit values.
    :rtype: Tuple[float, List[float]]
    """

    if n_partitions % 2 == 1:
        raise ValueError("Number of partitions must be even.")

    _, n_strategies = matrix_data.shape
    partitions = np.array_split(matrix_data, n_partitions)
    partition_combinations = list(combinations(partitions, len(partitions) // 2))

    results = Parallel(n_jobs=n_jobs)(
        delayed(performance_evaluation)(
            np.concatenate(train_parts, axis=0),
            np.concatenate([part for part in partitions if part not in train_parts], axis=0),
            n_strategies,
            metric,
            risk_free_return
        )
        for train_parts in partition_combinations
    )

    pbo = results[:, 0].mean(axis=0)
    logit_values = results[:, 1]

    return pbo, logit_values
```
